This Python notebook is a part of the Synapse Analyzer package. The purpose of the notebook is to take Excel files that were generated by Imaris, containing the statistics of all surfaces, and convert them into a simplified format that can be used by SynAnalyzer.ijm.

```python
import numpy as np
import pandas as pd
import glob
```

```python
# Generate a list of all xls files added by the user
files = glob.glob('*.xls')
print(files)
```

```python
# Iterate through the xls files, extract the relavent sheet, and reformat
for file in files:
    df = pd.read_excel(file,skiprows=1, sheet_name='Position')
    df = df.drop(columns=['Unit', 'Category', 'Collection', 'Time'])
    df.set_index('ID', inplace=True)
    df.to_csv(file.split('.x')[0]+'.csv')
```
