This Python notebook is a essentially a utility function for the SynAnalyzer pipeline. It allows the user to generate a metadata spreadsheet for batch analysis given a set of prep IDs. This file will be used in plotting the results from SynAnalyzer. 

```python
import pandas as pd
import os
import xlrd
import openpyxl
```

```python
#                *** WHAT TO ANALYZE // WHERE TO GET/STORE **
# Batch analysis directory (where the new metadata sheet will be stored)
batchID = 'SynAnalyzer_BclwTGAging'     
# Preps that are included in the analysis
preps = ['WSS_048','WSS_049']
# Metadata fields to grab
mdFields = ['SampleID','AnimalID','Sex','DOB','Group']

# Directories 
#   existing ones 
dirMain = '/Users/joyfranco/Dropbox (Partners HealthCare)/JF_Shared/Data/WSS/BatchAnalysis/'
dirPrep = '/Users/joyfranco/Dropbox (Partners HealthCare)/JF_Shared/Data/WSS/'
dirBA = dirMain+batchID+'/'
#   ones to make
dirBAMD = dirBA+'Metadata/'   
```

```python
if not os.path.exists(dirBAMD): os.mkdir(dirBAMD)
```

```python
#      *** BEGIN COMPILIATION PROCESS ***
dfMDAll = pd.DataFrame()
# Iterate through the preps and build the new metadata sheet
for prep in preps:
    # Get the metadata sheet
    dirSP = dirPrep+prep+'/'
    dirMD = dirSP+'Metadata/'
    fnEx = prep+'.Metadata.Samples.xlsx'
    
    # Load the sheet
    dfEF = pd.read_excel(dirMD+fnEx)

    # Add it to the main one
    dfMDAll = pd.concat([dfMDAll, dfEF[mdFields]])

dfMDAll.to_csv(dirBAMD+batchID+'.Metadata.Samples.csv') 
```
